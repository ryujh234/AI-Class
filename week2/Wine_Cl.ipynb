```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/AI/week2/wine.csv')

df
```

```python
y=df['Wine']
y.value_counts()
y
```

```python
X=df.drop('Wine',axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

wine_lr= LogisticRegression(random_state=0, max_iter=5000)
wine_lr.fit(X_train, y_train)

pred_lr = wine_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

wine_dt = DecisionTreeClassifier(random_state=0)
wine_dt.fit(X_train, y_train)

pred_dt = wine_dt.predict(X_test)

print ("\n--- DecisionTreeClassifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier

wine_rt = RandomForestClassifier(random_state=0)
wine_rt.fit(X_train, y_train)

pred_rt = wine_rt.predict(X_test)

print ("\n--- RandomForestClassifier ---")
print (accuracy_score(y_test, pred_rt))
print (confusion_matrix(y_test, pred_rt))
```

```python
from sklearn.neighbors import KNeighborsClassifier

wine_knn = KNeighborsClassifier(n_neighbors=5)
wine_knn.fit(X_train, y_train)

pred_knn = wine_knn.predict(X_test)

print ("\n--- KNeighborsClassifier ---")
print (accuracy_score(y_test, pred_knn))
print (confusion_matrix(y_test, pred_knn))
```

```python
# K값의 최적값을 찾기 위해 for문을 1~30까지 돌리고 그래프로 표현 Classifier에선 정확도가 가장 큰 값
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

k_list = []
acc_list = []

#최적 k 찾기
for k in range(1, 31):
    Wine_knn = KNeighborsClassifier(n_neighbors=k)
    Wine_knn.fit(X_train, y_train)

    pred_knn = Wine_knn.predict(X_test)

    acc = accuracy_score(y_test, pred_knn)

    k_list.append(k)
    acc_list.append(acc)

best_k = k_list[np.argmax(acc_list)]
best_acc = max(acc_list)

print("\n--- KNeighborsClassifier 최적화 결과 ---")
print("최적의 k:", best_k)
print("최고 정확도:", best_acc)

# 최적 k로 다시 학습
best_model = KNeighborsClassifier(n_neighbors=best_k)
best_model.fit(X_train, y_train)
pred_knn = best_model.predict(X_test)

print("\n--- KNeighborsClassifier ---")
print("정확도:", accuracy_score(y_test, pred_knn))
print("혼동행렬:\n", confusion_matrix(y_test, pred_knn))

# 그래프
plt.figure(figsize=(8, 5))
plt.plot(k_list, acc_list, marker='o')
plt.xlabel("k 값")
plt.ylabel("Accuracy")
plt.title("k 값에 따른 KNeighborsClassifier 성능")
plt.xticks(range(1, 31))
plt.grid(True)
plt.show()
```
