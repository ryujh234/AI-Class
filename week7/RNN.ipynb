# RNN(Recurrent Neural Network) - 순환 신경망

여러 개의 데이터가 순서대로 입력되었을 때 앞서 입력받은 데이터를 잠시 기억해 놓는 방법이다.
현재 시점의 입력만 사용하는 것이 아니라 이전 시점의 정보를 은닉 상태(hidden state)에 저장하고, 이 정보를 다음 시점으로 전달한다. 즉, RNN은 데이터를 순서대로 읽으면서 지금까지의 정보를 계속 업데이트하는 구조이다.

# 일반 신경망과 순환 신경망의 차이

RNN의 특징으로는 현재 입력만 보는 것이 아니라 이전에 처리한 정보도 함께 참고한다는 점이다. 이 때문에 RNN은 순서가 있는 데이터를 처리하는 데 적합하다.

RNN의 또 다른 특징은 같은 가중치를 여러 시점에서 반복적으로 사용한다는 점이다.예를 들어 길이가 10인 시퀀스를 처리할 때, RNN은 10개의 서로 다른 모델을 사용하는 것이 아니라 하나의 RNN 셀을 시간 방향으로 반복해서 사용한다. 이 구조 덕분에 입력 길이가 달라져도 같은 방식으로 처리할 수 있다.

RNN은 문제의 형태에 따라 여러 방식으로 사용할 수 있다.

Many-to-one 구조는 여러 개의 입력을 보고 하나의 결과를 출력하는 방식이다. 예를 들어 문장 전체를 읽고 긍정 또는 부정을 분류하는 감성 분석에 사용할 수 있다.

Many-to-many 구조는 입력 시퀀스마다 출력이 필요한 경우에 사용된다. 예를 들어 문장의 각 단어마다 품사를 예측하는 문제에 사용할 수 있다.

One-to-many 구조는 하나의 입력에서 여러 개의 출력을 생성하는 경우에 사용할 수 있다. 예를 들어 하나의 시작 단어를 바탕으로 문장을 생성하는 작업이 이에 해당한다.

# RNN의 장단점

RNN의 가장 큰 장점은 순서 정보를 반영할 수 있다는 점이다.

일반적인 Dense network는 입력의 순서를 직접 기억하지 못하지만, RNN은 이전 은닉 상태를 다음 시점으로 전달하기 때문에 문장이나 시계열 데이터처럼 순서가 중요한 데이터를 처리할 수 있다.

하지만 RNN에는 한계도 존재한다. 시퀀스가 길어질수록 앞쪽 정보가 뒤쪽까지 잘 전달되지 못하는 문제가 발생한다. 이를 장기 의존성 문제라고 한다.

또한 학습 과정에서 기울기 소실 문제가 발생할 수 있다. 기울기 소실이 발생하면 오래전에 입력된 정보가 모델 학습에 충분히 반영되지 못한다. 이 문제를 보완하기 위해 등장한 모델이 LSTM이다.
