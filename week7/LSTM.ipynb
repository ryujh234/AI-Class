# LSTM(Long Short-Term Memory) - 장단기 기억 신경망

순서가 있는 데이터를 처리할 수 있지만, 시퀀스가 길어질수록 앞쪽 정보를 잘 기억하지 못한다는 한계가 있다(그래디언트 소실 문제).


이러한 RNN의 단점을 보완하기 위해 나온 것이 LSTM이다.

LSTM은 Long Short-Term Memory의 약자로, 중요한 정보는 오래 기억하고 불필요한 정보는 버리도록 설계된 순환 신경망이다.

일반 RNN이 은닉 상태 하나에 많은 정보를 담으려고 한다면, LSTM은 셀 상태(cell state)라는 별도의 기억 공간을 사용한다. 셀 상태는 장기 기억의 역할을 하며, 게이트 구조를 통해 어떤 정보를 유지하고 어떤 정보를 버릴지 조절한다.

LSTM의 핵심 구성 요소는 크게 네 가지로 볼 수 있다.

첫 번째는 셀 상태이다. 셀 상태는 LSTM에서 장기 기억을 담당하는 통로이다. 중요한 정보가 시퀀스의 뒤쪽까지 전달될 수 있도록 도와준다.

두 번째는 Forget Gate이다. Forget Gate는 이전 정보 중에서 무엇을 버릴지 결정한다. 더 이상 필요 없는 정보는 약하게 만들고, 중요한 정보는 유지한다.

세 번째는 Input Gate이다. Input Gate는 새로 들어온 정보 중에서 무엇을 기억할지 결정한다. 모든 입력을 저장하는 것이 아니라 현재 문제 해결에 필요한 정보만 선택적으로 저장한다.

네 번째는 Output Gate이다. Output Gate는 현재 시점에서 어떤 정보를 출력할지 결정한다. LSTM이 내부에 저장한 정보 중 현재 출력에 필요한 부분만 밖으로 내보낸다.

LSTM은 다음과 같은 흐름으로 동작한다.

1. 이전 셀 상태에서 불필요한 정보를 잊는다.
2. 현재 입력에서 새롭게 저장할 정보를 선택한다.
3. 셀 상태를 업데이트한다.
4. 현재 시점의 출력값을 만든다.

이 과정을 통해 LSTM은 긴 시퀀스에서도 중요한 정보를 더 오래 유지할 수 있다.

# RNN과 LSTM의 차이

RNN은 주로 은닉 상태를 이용해 이전 정보를 전달한다. 구조가 비교적 단순하기 때문에 이해하기 쉽고 계산도 상대적으로 가볍다. 하지만 시퀀스가 길어지면 앞쪽 정보를 오래 유지하기 어렵다.

LSTM은 은닉 상태뿐만 아니라 셀 상태를 함께 사용한다. 셀 상태는 장기 기억을 담당하고, 여러 게이트가 정보의 흐름을 조절한다. 따라서 일반 RNN보다 긴 문장이나 긴 시계열 데이터를 처리하는 데 더 적합하다.

다만 LSTM은 RNN보다 구조가 복잡하기 때문에 계산량이 더 많고 학습 시간이 더 오래 걸릴 수 있다.

RNN은 이전 정보를 모두 넘긴다면 LSTM의 경우 어떤 정보는 장기기억으로 처리하여 cell state에 저장되고, 어떤 정보는 forget gate로 제거하고, input gate로 새로운 정보 중에서 어떤 정보를 저장할지 분류하고, output gate로 어떤 정보를 출력할지를 정하는 방식이다.
