# CNN(Convolutional Neural Networks) - 합성곱 신경망

Dense,  fully connected network 를 하게 될 경우 모든 입력을 단순한 숫자 나열로 보기 때문에 이미지의 공간적인 특성이 사라지게 된다.

이를 해결하기 위해 나온것이 CNN-합성곱 신경망 이다.
합성곱이란 원본 이미지에 작은 커널(필터)을 적용하여 새로운 특징을 만들어내는 과정이다.
커널마다 가로선, 세로선, 모서리, 곡선 등의 특징을 감지하고 이를통해 특징맵을 만들어 특징맵을 통하여 이미지에서 어떤 특징이 어디서 강하게 존재하는지 나타낸다.
CNN 초기 단계에서는 비교적 단순한 특징을 찾고, 깊은층으로 갈수록 눈, 코, 얼굴, 바퀴, 동물의 귀처럼 복잡한 특징을 찾게 된다.

CNN에서 필터는 사람이 직접 정해주는 것이 아니라, 학습을 통해 최적화된다. 처음 필터의 값은 무작위로 설정되지만, 학습 데이터와 정답을 이용해 반복적으로 훈련하면서 이미지 분류에 유용한 값으로 바뀐다.필터 크기가 커지면 한 번에 더 넓은 영역을 볼 수 있지만, 그만큼 학습해야 할 가중치도 증가한다. 또한 필터의 개수가 많아질수록 다양한 특징을 추출할 수 있지만, 생성되는 특징맵의 개수도 많아지고 계산량도 증가한다.
이 문제를 해결하기 위해 CNN에서는 보통 합성곱 연산 뒤에 풀링, Pooling 연산을 사용한다.

# 풀링(Pooling)

풀링은 입력 데이터의 크기를 줄이는 과정이다.
풀링의 주요 기능은 크게 두 가지로 정보 요약과 이동에 대한 불변성이다.
풀링은 넓은 영역의 정보를 하나의 값으로 압축한다. 따라서 불필요한 세부 정보는 줄이고 중요한 특징만 남긴다.
이미지 속 물체가 약간 이동하더라도 풀링을 사용하면 비슷한 결과를 얻을 수 있다.

- 최대 풀링(Max Pooling)

정해진 영역 안에서 가장 큰 값을 선택하는 방식으로 윈도우를 움직이면서 그 안에 있는 숫자 중 가장 큰 값만 출력하는 연산이다.

- 평균 풀링(Average Pooling)

정해진 영역 안의 값들의 평균을 구하는 방식으로 전체적인 정보를 부드럽게 요약하는 데 적합하다. 최대 풀링이 가장 강한 특징을 강조한다면, 평균 풀링은 영역 전체의 경향을 반영한다고 볼 수 있다.

# 보폭(Stride)

필터가 한 번에 이동하는 거리를 의미한다. stride가 1이면 커널을 한 번에 1픽셀씩 이동하고, stride가 2이면 하나씩 건너뛰면서 적용한다. Stride가 작으면 필터가 촘촘하게 이동하므로 더 많은 정보를 얻을 수 있다. 하지만 출력 특징맵의 크기가 커지고 계산량도 많아진다. 반대로 stride가 크면 필터가 넓게 이동하므로 출력 크기가 줄어들고 계산량도 감소하지만, 일부 세부 정보가 손실될 수 있다.

# 패딩(Padding)

이미지의 가장자리를 처리하기 위한 방법으로 대표적으로 Valid와 Same 방식이 있다. Valid는 패딩을 사용하지 않고 커널을 입력 이미지 안에서만 움직이는 방식이다. 이 경우 출력 크기가 입력보다 작아진다. Same은 입력 이미지 주변을 0 또는 특정 값으로 채워 출력 크기를 입력과 비슷하게 유지하는 방식이다.

- Zero Padding

이미지 주위를 0으로 둘러주는 과정으로 입출력 이미지의 크기를 유지할 수 있다.

전체 흐름

입력 이미지 → 합성곱 층 → 활성화 함수 → 풀링 층 → 합성곱 층 반복 → 평탄화 → 완전연결 층 → 출력
